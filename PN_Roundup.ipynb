# PuckNight Roundup - 2022 Winter Olympics
## Quarter-Final Matchups
## Beijing, China - February 9-20,  2022
### The 2022 Winter Olympic Hockey Tournament 

```python
league = 'OG'
season = '2021-22'
from pucknight import pn_queries
import pandas as pd
from datetime import datetime
stand = pn_queries.league_standings(league, season)
scoring = pn_queries.league_scoring(league, season)
goalies = pn_queries.league_goalies(league, season)
scoring['pts']= scoring['g']+scoring['a']

def age(d1, d2):
    d1 = datetime.strptime(d1, "%Y-%m-%d")
    d2 = datetime.strptime(d2, "%Y-%m-%d")
    return round(abs((d2 - d1).days/365),1)
```

## Round-Robin Standings

```python
stand.groupby(['division','team']).agg({'gp':'max','points':'max'})\
.sort_values(['division','points'], ascending=[True, False])
```

```python
scoring[['team','playername','gp','pts','g','a','nation','age_curr','playerid']].sort_values('pts', ascending=False)\
    .iloc[:12]
```

```python
goalies[goalies['gp']>0][['team','playername','gp','gaa','sv_prc','age_curr','playerid']].sort_values(['gp','gaa']
                                                                                     , ascending=[False,True])
```

```python
roster_info = pd.DataFrame()
for pid in list(scoring.playerid):
    p_info = pn_queries.player_info(pid)
    roster_info = pd.concat([roster_info, p_info])
for pid in list(goalies.playerid):
    p_info = pn_queries.player_info(pid)
    roster_info = pd.concat([roster_info, p_info])
s_info = roster_info.merge(scoring[['playerid','team','gp']], on=['playerid'])
g_info = roster_info.merge(goalies[['playerid','team','gp']], on=['playerid'])
roster_info = pd.concat([s_info, g_info])
        
plr_careers = pd.DataFrame()
for pid in list(scoring.playerid):
    p_career = pn_queries.player_career(pid)
    plr_careers = pd.concat([plr_careers, p_career])    
for pid in list(goalies.playerid):
    p_career = pn_queries.player_career(pid)
    plr_careers = pd.concat([plr_careers, p_career])
s_careers = plr_careers.merge(scoring[['playerid','team','gp']], on=['playerid'], suffixes=['_car','_cur'])
g_careers = plr_careers.merge(goalies[['playerid','team','gp']], on=['playerid'], suffixes=['_car','_cur'])
plr_careers = pd.concat([s_careers, g_careers])    
    
plr_car_tot = pd.DataFrame()
for pid in list(scoring.playerid):
    p_car_sum = pn_queries.player_summed_career(pid)
    plr_car_tot = pd.concat([plr_car_tot,p_car_sum])
for pid in list(goalies.playerid):
    p_car_sum = pn_queries.player_summed_career(pid)
    plr_car_tot = pd.concat([plr_car_tot,p_car_sum])    
s_car_tot = plr_careers.merge(scoring[['playerid','team','gp']], on=['playerid'], suffixes=['_car','_cur'])
g_car_tot = plr_careers.merge(goalies[['playerid','team','gp']], on=['playerid'], suffixes=['_car','_cur'])
plr_car_tot = pd.concat([s_careers, g_careers])    
    
plr_awards = pd.DataFrame()
for pid in list(scoring.playerid):
    plr_aw = pn_queries.awards(pid)
    plr_awards = pd.concat([plr_awards,plr_aw])
for pid in list(goalies.playerid):
    plr_aw = pn_queries.awards(pid)
    plr_awards = pd.concat([plr_awards,plr_aw]) 
plr_awards = plr_awards.merge(plr_careers, on=['playerid','league','season'])
s_awards = plr_awards.merge(scoring[['playerid','team','gp']], on=['playerid'], suffixes=['_aw','_cur'])
g_awards = plr_awards.merge(goalies[['playerid','team','gp']], on=['playerid'], suffixes=['_aw','_cur'])
plr_awards = pd.concat([s_awards, g_awards])        

roster_info['age'] = roster_info['dob'].\
apply(lambda row: age(str(row),'2022-02-04'))
```

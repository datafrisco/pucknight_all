# PuckNight Preview - 2022 Olympic Games
## Beijing, China - February 2022

```python
from pucknight import pn_queries
import pandas as pd
from datetime import datetime
league = 'OG'
season = '2021-22'
```

```python
stand = pn_queries.league_standings(league, season)
scoring = pn_queries.league_scoring(league, season)
goalies = pn_queries.league_goalies(league, season)

def age(d1, d2):
    d1 = datetime.strptime(d1, "%Y-%m-%d")
    d2 = datetime.strptime(d2, "%Y-%m-%d")
    return round(abs((d2 - d1).days/365),1)
```

```python
stand[['division','team','gp']].sort_values(['division','team'])
```

```python
team = 'USA'
roster_list = list(scoring[scoring['team']==team]['playerid'])
goalie_list = list(goalies[goalies['team']==team]['playerid'])

team_info = pd.DataFrame()
for pid in roster_list:
    p_info = pn_queries.player_info(pid)
    team_info = pd.concat([team_info, p_info])
for pid in goalie_list:
    p_info = pn_queries.player_info(pid)
    team_info = pd.concat([team_info, p_info])
        
team_careers = pd.DataFrame()
for pid in roster_list:
    p_career = pn_queries.player_career(pid)
    team_careers = pd.concat([team_careers, p_career])    
for pid in goalie_list:
    p_career = pn_queries.player_career(pid)
    team_careers = pd.concat([team_careers, p_career])    
    
team_car_tot = pd.DataFrame()
for pid in roster_list:
    p_car_sum = pn_queries.player_summed_career(pid)
    team_car_tot = pd.concat([team_car_tot,p_car_sum])
for pid in goalie_list:
    p_car_sum = pn_queries.player_summed_career(pid)
    team_car_tot = pd.concat([team_car_tot,p_car_sum])    
    
team_awards = pd.DataFrame()
for pid in roster_list:
    plr_aw = pn_queries.awards(pid)
    team_awards = pd.concat([team_awards,plr_aw])
for pid in goalie_list:
    plr_aw = pn_queries.awards(pid)
    team_awards = pd.concat([team_awards,plr_aw])    

team_info['age'] = team_info['dob'].\
apply(lambda row: age(str(row),'2022-02-04'))

szn_careers = team_careers[(team_careers['season']=='2021-22') & (team_careers['league'] != 'OG')]\
[['playerid','league','team','gp','g_gaa','tp_svpct']]
             
roster_szn = team_info.merge(szn_careers, on='playerid')
```

## Team USA - Roster and 2021-22 Club Team Stats

```python
roster_szn.groupby(['playername','age','position','place_of_birth','draft_year','overall_pick','league']).\
agg({'team':'max','gp':'max','g_gaa':'max','tp_svpct':'max'}).\
sort_values(['age','playername','position'])
```
